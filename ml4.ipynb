Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point 
x=2

```python
curx = 2
rate = 0.01
precision = 0.000001
pre_step_size = 1
iter = 0
max_iter= 10000
df = lambda x : 2*(x+3)
arr = []
while pre_step_size > precision and max_iter> iter:
    prex = curx
    curx = curx - rate * df(prex)
    pre_step_size = abs(curx-prex)
    iter = iter+1
    arr.append(curx)
    print('iteraion : ',iter,' value : ',curx)
```

```python
print('iteraion : ',iter,' value : ',curx)
print('min : ',min(arr))
```
